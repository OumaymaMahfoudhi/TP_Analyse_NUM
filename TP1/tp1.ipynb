### La méthode de dichotomie 

```python
f=lambda x: x**3+x**2-3*x-3
def dichotomie (f , a , b , epsilon ):
            g , d = min (a , b ) , max (a , b )
            valg , vald = f ( g ) , f ( d )
            n = 0
            while d - g > epsilon :
                n += 1
                m = ( g + d ) / 2
                valm = f ( m )
                if valg * valm < 0 :
                    d = m
                    vald = valm
                else :
                    g = m
                    valg = valm
            print (" nombre d'itérations nécessaires :", n )
            return ( g + d ) / 2
```

- $\alpha_1 \in ]-2;-1.5[$ 
- $\alpha_2 \in ]-1.5;-0.5[$ 
- $\alpha_3 \in ]1.5;2[$

```python
f=lambda x:  x**3+4*x**2-10 # la fonction f
g1=lambda x: 1/2*sqrt(10-x**3)
g2=lambda x: sqrt(10/(x+4))
g3=lambda x: x-x**3-4*x**2 + 10


from math import *

def point_fixe(g, x0, epsilon, n_itermax):
  """
    Retourne un tableau avec tous les points calculés ou une erreur 
  """
  itr = 0
  xp = {}
  x = 0

  while (itr < n_itermax):
    try:
      x = g(x0)
    except OverflowError:
      break

    xp[itr] = x0
    if abs(x - x0) <= epsilon:
      return xp

    x0 = x
    itr += 1

  return "Aucun point fixe"

def print_table(xp):
  """
    Affiche un tableau avec toutes les valeurs ou l'erreur
  """
  s = ""
  if isinstance(xp, str):
    s = xp
  else:
    for key, value in xp.items():
      s += "n = {}\t\tx = {}\n".format(key, value)
  print(s)
   

```

```python
def newton(fonction,derivee,a,e):
    delta = 1
    while delta > e:
        x = -fonction(a)/derivee(a) + a
        delta = abs(x - a)
        a = x
        
    return x , delta
        
print( newton(lambda x: x**2+2*x+1 , lambda x: 2*x+2 , 0 , 0.001) )
```
